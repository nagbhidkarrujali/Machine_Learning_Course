### Calculate daily gains
$$ gain = {(Close Price - Open Price) * 100 \over Open Price} $$

```python
hindustan['Gain'] = ((hindustan['Close Price']-hindustan['Open Price'])*100/hindustan['Open Price'])
hindustan[:5]
```

```python
itc_ltd['Gain'] = ((itc_ltd['Close Price']-itc_ltd['Open Price'])*100/itc_ltd['Open Price'])
itc_ltd[:5]
```

```python
lt_foods['Gain'] = ((lt_foods['Close Price']-lt_foods['Open Price'])*100/lt_foods['Open Price'])
lt_foods[:5]
```

```python
nes['Gain'] = ((nes['Close Price']-nes['Open Price'])*100/nes['Open Price'])
nes[:5]
```

```python
parle['Gain'] = ((parle['Close Price']-parle['Open Price'])*100/parle['Open Price'])
parle[:5]
```

## Confidence Interval
### Value of Risk at 95% - Hindustan Unilever

```python
hindustan_ci_95 = stats.norm.interval(0.95,
                                      loc=hindustan.Gain.mean(),
                                      scale=hindustan.Gain.std())

hindustan_ci_95
```

### Value of Risk at 95% - ITC LTD

```python
itc_ltd_ci_95 = stats.norm.interval(0.95,
                                      loc=itc_ltd.Gain.mean(),
                                      scale=itc_ltd.Gain.std())

itc_ltd_ci_95
```

### Value of Risk at 95% - LT FOODS

```python
lt_foods_ci_95 = stats.norm.interval(0.95,
                                      loc=lt_foods.Gain.mean(),
                                      scale=lt_foods.Gain.std())

lt_foods_ci_95
```

### Value of Risk at 95% - Nestle

```python
nes_ci_95 = stats.norm.interval(0.95,
                                      loc=nes.Gain.mean(),
                                      scale=nes.Gain.std())

nes_ci_95
```

### Value of Risk at 95% - Parle

```python
parle_ci_95 = stats.norm.interval(0.95,
                                      loc=parle.Gain.mean(),
                                      scale=parle.Gain.std())

parle_ci_95
```

### What is the probability that the stock will make a gain of 4% or more?

#### HINDUSTAN UNILEVER - probability of a gain of 4%

```python
1 - stats.norm.cdf(4.0,
                    loc=hindustan.Gain.mean(),
                    scale=hindustan.Gain.std())
```

#### ITC LTD - probability of a gain of 4%

```python
1 - stats.norm.cdf(4.0,
                    loc=itc_ltd.Gain.mean(),
                    scale=itc_ltd.Gain.std())
```

#### LT FOODS - probability of a gain of 4%

```python
1 - stats.norm.cdf(4.0,
                    loc=lt_foods.Gain.mean(),
                    scale=lt_foods.Gain.std())
```

#### NESTLE - probability of a gain of 4%

```python
1 - stats.norm.cdf(4.0,
                    loc=nes.Gain.mean(),
                    scale=nes.Gain.std())
```

#### PARLE - probability of a gain of 4%

```python
1 - stats.norm.cdf(4.0,
                    loc=parle.Gain.mean(),
                    scale=parle.Gain.std())
```
