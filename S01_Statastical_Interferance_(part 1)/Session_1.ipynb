```python
import pandas as pd
import matplotlib as mplot
import matplotlib.pyplot as plt
import seaborn as sn
from scipy import stats
```

### Load Hisdustan_Unilever Data

```python
hindustan = pd.read_csv('E:\ML_course\practice\S1_Statastical_Interferance_(part 1)\Hindustan_Unilever.csv', parse_dates=['Date'] )
hindustan = hindustan.set_index(['Date'], drop=True)
hindustan.sort_index(ascending =True, inplace= True)
hindustan = hindustan[['Open Price','Close Price']]
hindustan[:5]
```

### Load ITC LTD Data

```python
itc_ltd = pd.read_csv('E:\ML_course\practice\S1_Statastical_Interferance_(part 1)\itc_ltd.csv', parse_dates=['Date'] )
itc_ltd = itc_ltd.set_index(['Date'], drop= True)
itc_ltd.sort_index(ascending=True , inplace=True)
itc_ltd = itc_ltd[['Open Price','Close Price']]
itc_ltd[:5]
```

### Load LT Foods Data

```python
lt_foods = pd.read_csv('E:\ML_course\practice\S1_Statastical_Interferance_(part 1)\lt_foods.csv',parse_dates=['Date'] )
lt_foods = lt_foods.set_index(['Date'], drop = True)
lt_foods.sort_index(ascending=True , inplace=True)
lt_foods = lt_foods[['Open Price','Close Price']]
lt_foods[:5]
```

### Load Nestle Data

```python
nes = pd.read_csv('E:\ML_course\practice\S1_Statastical_Interferance_(part 1)\_nestle.csv',parse_dates=['Date'] )
nes = nes.set_index(['Date'], drop = True)
nes.sort_index(ascending=True , inplace=True)
nes = nes[['Open Price','Close Price']]
nes[:5]
```

### Load Parle Data

```python
parle = pd.read_csv('E:\ML_course\practice\S1_Statastical_Interferance_(part 1)\Parle.csv',parse_dates=['Date'] )
parle = parle.set_index(['Date'], drop = True)
parle.sort_index(ascending=True , inplace=True)
parle = parle[['Open Price','Close Price']]
parle[:5]
```

### Calculate daily gains
$$ gain = {(Close Price - Open Price) * 100 \over Open Price} $$

```python
hindustan['Gain'] = ((hindustan['Close Price']-hindustan['Open Price'])*100/hindustan['Open Price'])
# hindustan[:5]
```

```python
itc_ltd['Gain'] = ((itc_ltd['Close Price']-itc_ltd['Open Price'])*100/itc_ltd['Open Price'])
# itc_ltd[:5]
```

```python
lt_foods['Gain'] = ((lt_foods['Close Price']-lt_foods['Open Price'])*100/lt_foods['Open Price'])
# lt_foods[:5]
```

```python
nes['Gain'] = ((nes['Close Price']-nes['Open Price'])*100/nes['Open Price'])
# nes[:5]
```

```python
parle['Gain'] = ((parle['Close Price']-parle['Open Price'])*100/parle['Open Price'])
# parle[:5]
```

### Plotting Distribution Plots

```python
plt.figure(figsize=(12,4))
sn.kdeplot(hindustan.Gain, label='hindustan_unilever')
sn.kdeplot(itc_ltd.Gain, label='ITC_LTD')
sn.kdeplot(lt_foods.Gain, label='LT_FOODS')
sn.kdeplot(nes.Gain, label='NESTLE')
sn.kdeplot(parle.Gain, label='PARLE')
plt.title("Volatility of Stocks")
plt.legend()
```

### Calculate Mean

```python
print("Hindustan : " ,hindustan.Gain.mean())
print("itc_ltd : " ,itc_ltd.Gain.mean())
print("Lt_foods : " ,lt_foods.Gain.mean())
print("nestle : " ,nes.Gain.mean())
print("Parle : " ,parle.Gain.mean())
```

### Standard Deviation

```python
print("Hindustan : " ,hindustan.Gain.std())
print("itc_ltd : " ,itc_ltd.Gain.std())
print("Lt_foods : " ,lt_foods.Gain.std())
print("nestle : " ,nes.Gain.std())
print("Parle : " ,parle.Gain.std())
```

## Confidence Interval
- Confidence interval is a range of values, bounded above and below the mean value.

- It is the probability that a parameter will fall between a set of values for a certain proportion of times. 

- Most often used confidence intervals are either 90% or 95% or 99%.
### Value of Risk at 95%

```python
hindustan_ci_95 = stats.norm.interval(0.95,
                                      loc=hindustan.Gain.mean(),
                                      scale=hindustan.Gain.std())
itc_ltd_ci_95 = stats.norm.interval(0.95,
                                      loc=itc_ltd.Gain.mean(),
                                      scale=itc_ltd.Gain.std())
lt_foods_ci_95 = stats.norm.interval(0.95,
                                      loc=lt_foods.Gain.mean(),
                                      scale=lt_foods.Gain.std())
nes_ci_95 = stats.norm.interval(0.95,
                                      loc=nes.Gain.mean(),
                                      scale=nes.Gain.std())
parle_ci_95 = stats.norm.interval(0.95,
                                      loc=parle.Gain.mean(),
                                      scale=parle.Gain.std())

print("Hindustan : " ,hindustan_ci_95)
print("itc_ltd : " ,itc_ltd_ci_95)
print("Lt_foods : " ,lt_foods_ci_95)
print("nestle : " ,nes_ci_95)
print("Parle : " ,parle_ci_95)
```

### What is the probability that the stock will make a gain of 4% or more?

```python
print("Hindustan : ",1 - stats.norm.cdf(4.0,
                                        loc=hindustan.Gain.mean(),
                                        scale=hindustan.Gain.std()))
print("ITC LTC : ",1 - stats.norm.cdf(4.0,
                                        loc=itc_ltd.Gain.mean(),
                                        scale=itc_ltd.Gain.std()))
print("LT foods : ",1 - stats.norm.cdf(4.0,
                                        loc=lt_foods.Gain.mean(),
                                        scale=lt_foods.Gain.std()))
print("Nestle : ",1 - stats.norm.cdf(4.0,
                                        loc=nes.Gain.mean(),
                                        scale=nes.Gain.std()))
print("Parle : ",1 - stats.norm.cdf(4.0,
                                        loc=parle.Gain.mean(),
                                        scale=parle.Gain.std()))
```

### Plot the daily gain or loss of all the stocks in one plot.

```python
plt.figure(figsize=(12,5))
plt.plot(hindustan.Gain, label='hindustan_unilever')
plt.plot(itc_ltd.Gain, label='ITC_LTD')
plt.plot(lt_foods.Gain, label='LT_FOODS')
plt.plot(nes.Gain, label='NESTLE')
plt.plot(parle.Gain, label='PARLE')
plt.legend()
```
